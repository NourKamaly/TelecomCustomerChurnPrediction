```python
def get_feature_percentage(df):
    size = df.shape[0]
    for column in df.columns:
        ctr = 0
        if df[column].dtype == np.dtype(object):
            for index in df[column].value_counts().index:
                    percentage = (int(df[column].value_counts()[ctr])/size)*100
                    print("Feature {feature} {index} represents {percentage} of the data".format(feature= column,index = index,percentage =percentage))
                    ctr += 1
            print('----------')
```
