```python
features = df.drop(columns = ['Churn'])
labels = df['Churn']
x_train,x_validation, y_train,y_validation = train_test_split(features, labels, test_size = 0.2, stratify = labels, random_state=42)
x_train['Churn'], x_validation['Churn'] = y_train, y_validation
x_validation.to_csv('../0. Dataset/Validation set.csv',index=False)
x_train.to_csv('../0. Dataset/Training set after splitting.csv',index=False)
```

```python
df = pd.read_csv('../0. Dataset/Training set after splitting.csv')
df.drop(columns = ['CustomerID'],axis=1,inplace=True)
```

```python
def get_feature_percentage(df):
    size = df.shape[0]
    for column in df.columns:
        ctr = 0
        if df[column].dtype == np.dtype(object):
            for index in df[column].value_counts().index:
                    percentage = (int(df[column].value_counts()[ctr])/size)*100
                    print("Feature {feature} {index} represents {percentage} of the data".format(feature= column,index = index,percentage =percentage))
                    ctr += 1
            print('----------')
```

```python
get_feature_percentage(df)
```
